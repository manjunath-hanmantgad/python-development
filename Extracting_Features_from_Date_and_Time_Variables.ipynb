```python
import pandas as pd
import numpy as np
```

```python
#create 20 datetime values beginning from 2019-03-05 at midnight, followed by increments of 1 day. Then, we must capture the values in a DataFrame and display the top five rows:

range_date = pd.date_range("2019-03-05", periods=20, freq="D")
data = pd.DataFrame({"date":range_date})
data.head()
```

```python
data["date"].dtypes
```

```python
#extract the year part of the date in a new column and display the top five rows of the resulting DataFrame:
data["year"] = data["date"].dt.year
data.head()
```

```python
#extract the quarter out of the date into a new column and display the top five rows:
data["quarter"] = data["date"].dt.quarter
```

```python
data.head()
```

```python
#extract the month part of the date in a new column and display the top five rows of the DataFrame
data["month"] = data["date"].dt.month
data.head()
```

```python
#extract the week number (in a year) that the date corresponds to 
data["week_number"] = data["date"].dt.isocalendar().week
data.head()
```

```python
# extract the day of the month, which can take values between 1 and 31, and capture it in a new column. Then, we will display the top rows of the DataFrame:

data["day_of_month"] = data["date"].dt.day
data[["date", "day_of_month"]]
data.head()
```

```python
#extract the day of the week, with values between 0 and 6 (from Monday to Sunday), in a new column, then display the top rows:
data["day_of_week"] = data["date"].dt.dayofweek
data[["date", "day_of_week"]]
data.head()
```

```python
#lets create a binary variable that indicates whether the date was a weekend. Then, we will display the DataFrames top rows:

data["is_weekend"] = (
    data["date"].dt.dayofweek >= 5).astype(int) #assuming friday as weekend
data[["date","day_of_week","is_weekend"]].head()
```
