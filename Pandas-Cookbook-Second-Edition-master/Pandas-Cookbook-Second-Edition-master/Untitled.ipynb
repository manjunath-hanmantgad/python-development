#### Comparing continuous values across categories

We will look at mileage numbers in different brands: Ford, Honda, Tesla, and BMW.

- Make a mask for the brands we want and then use a group by operation to look at the mean and standard deviation for the city08 column for each group of cars:

```python
mask = fuel.make.isin(["Ford", "Honda", "Tesla", "BMW"])
fuel[mask].groupby('make').city08.agg(['mean','std'])
```
