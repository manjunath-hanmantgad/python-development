There is no need to use 64 bits for the RELAFFIL column as it contains only 0 or 1. Let's convert this column to an 8-bit (1 byte) integer with the .astype method:

```python
college['RELAFFIL']= college["RELAFFIL"].astype('int8')
```

```python
college.memory_usage(deep=True) #RELAFFIL -> 7535
# STABBR -> 444565
```

```python
college.dtypes # RELAFFIL -> int8
```

To save even more memory, you will want to consider changing object data types to categorical if they have a reasonably low cardinality (number of unique values). Let's first check the number of unique values for both the object columns:

```python
college.select_dtypes(include=["object"]).nunique()
```

```python
college.select_dtypes(include=["object"]).unique()
# bcoz unique applies on columns of dataframe
# nunique applies on entire dataframe
```

```python
college.select_dtypes(include=["object"]).nunique() / college.select_dtypes(include=["object"]).nunique().sum()
```

```python
# STABBR                0.004630 
# less than 1 % unique values

# so convert to catgeorical

college["STABBR"] = college["STABBR"].astype("category")
college.dtypes
```

```python
college.STABBR.memory_usage(deep=True) # 13248
```

#### Selecting the smallest of the largest

* This recipe can be used to create catchy news headlines such as Out of the Top 100 Universities, These 5 have the Lowest Tuition, or From the Top 50 Cities to Live, these 10 are the Most Affordable.

* During analysis, it is possible that you will first need to find a grouping of data that contains the top n values in a single column and, from this subset, find the bottom m values based on a different column.

Read in the movie dataset, and select the columns: movie_title, imdb_score, and budget:

```python
movies = pd.read_csv('data/movie.csv')
movie2 = movies[["movie_title","imdb_score","budget"]]
```

```python
movie2
```

Use the .nlargest method to select the top 100 movies by imdb_score

```python
movie2.nlargest(n=100,columns='imdb_score') 
#Signature: movie2.nlargest(n: 'int', columns: 'IndexLabel', keep: 'str' = 'first') -> 'DataFrame'
```

Chain the .nsmallest method to return the five lowest budget films among those with a top 100 score:

```python
movie2.nlargest(100,'imdb_score').nsmallest(5,"budget")
```

#### Selecting the largest of each group by sorting

find the highest-rated film of each year.

```python
movie3 = movies[["movie_title","imdb_score","title_year"]]
```

```python
movie3
```

Use the .sort_values method to sort the DataFrame by title_year

```python
movie3.sort_values(by='title_year', ascending=True)
```

```python
# sort multiple columns
movie3.sort_values(by=['title_year','imdb_score'], ascending=True)
```

```python
movie3.sort_values(by=['imdb_score','title_year'], ascending=False)
```

use the .drop_duplicates method to keep only the first row of every year

```python
movie3.sort_values(by=['title_year','imdb_score'], ascending=True).drop_duplicates(subset='title_year')
```

##### The default behavior of the .drop_duplicates method is to keep the first occurrence of each unique row, which would not drop any rows as each row is unique. However, the subset parameter alters it to only consider the column (or list of columns) given to it. In this example, only one row for each year will be returned. As we sorted by year and score in the last step, the highest-scoring movie for each year is what we get.
