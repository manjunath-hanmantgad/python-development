.iloc for lists

.loc for dictionaries

##### given a Series s, you can select data in the following ways: s[item] and s.loc[item]. The first performs the index operation directly on the Series. The second performs the index operation on the .loc attribute.

Read in the college dataset with the institution name as the index, and select a single column as a Series using an index operation:

```python
college = pd.read_csv('data/college.csv', index_col="INSTNM")
```

```python
college
```

```python
city = college["CITY"]
```

```python
city
```

```python
#Pull out a scalar value from the Series directly
city['National Personal Training Institute of Cleveland']
```

```python
# Pull out a scalar value using the .loc attribute by name
city.loc['National Personal Training Institute of Cleveland']
```

```python
city.iloc[100]
```

```python
>>> city[
...     [
...         "Alabama A & M University",
...         "Alabama State University",
...     ]
... ]
```

```python
>>> city.loc[
...     [
...         "Alabama A & M University",
...         "Alabama State University",
...     ]
... ]
```

```python
city.iloc[[0,4]]
```

```python
city.iloc[0,4] # means ek bari mai ek hi index position milega
# do index position chahye toh usko list mai dalo 
# [[index pos_a , index_pos_b]]
```

```python
>>> city[
...     "Alabama A & M University":"Alabama State University"
... ]
```

```python
city[:5]
```

```python
city.iloc[:5]
```

```python
city[:5] == city.iloc[:5]
```

```python
# select entire row from dataframe

college.iloc[60]
```

To select a disjointed set of rows as a DataFrame, pass a list of integers to .iloc:


```python
college.iloc[[10,20,30]]
```

```python
# get row and columns both

college.iloc[:1,:2] # 1st row and first 2 columns ka value milega
```

```python
college.head(3)
```

```python
# 3rd row ka satvrmid column ka value
college.iloc[:, [3,6]].head()
```

```python
college.iloc[:1,6]
```

```python
college.iloc[:1,[6]] ## 1st row ka satvrmid column ka value
```

```python
college.iloc[[3],[6]] ## 3rd row ka satvrmid column ka value
```

Select disjointed rows and columns

```python
college.iloc[[100,200],[0,1]] # 100 row and 200 row ka 1st and 2nd column
```

Slice the rows and select a single column:

```python
college.iloc[1:2,5]
```

```python
college.iloc[1:3,5]
```

#### Slicing lexicographically

Specifically, .loc allows you to select all rows with an index lexicographically using slice notation. This only works if the index is sorted.

```python
college.loc["Sp":"Su"]
```

```python
college = college.sort_index()
```

```python
college.loc["Sp":"Su"]
```
