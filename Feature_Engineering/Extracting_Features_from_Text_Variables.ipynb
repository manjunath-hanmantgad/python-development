### Implementing term frequency-inverse document frequency

```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
data = fetch_20newsgroups(subset='train')
df = pd.DataFrame(data.data, columns=['text'])

#remove punctuation and numbers from the text variable:
df['text'] = df['text'].str.replace('[^\w\s]','', regex=True).str.replace('\d+','',regex=True)

#set up TfidfVectorizer() from scikit-learn so that, 
#before creating the TF-IDF metrics, 
#it puts all text in lowercase, removes stop words, and retains words that appear in at least 5% of the text pieces:

vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words='english',
    ngram_range=(1, 1),
    min_df=0.05)

# fit TfidfVectorizer() so that it learns which words should be introduced as columns of the TF-IDF matrix:
vectorizer.fit(df['text'])
# create tf-idf matrix 
X = vectorizer.transform(df['text'])

#capture the TF-IDF matrix in a DataFrame with the corresponding feature names:
tfidf = pd.DataFrame(
    X.toarray(),
    columns = vectorizer.get_feature_names_out()
)
```

### Stemming

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.datasets import fetch_20newsgroups
data = fetch_20newsgroups(subset='train')
df = pd.DataFrame(data.data, columns=['text'])

# remove punctuation 
df["text"] = df['text'].str.replace('[^\w\s]','')

# remove characters that are numbers, leaving only letters
df['text'] = df['text'].str.replace('\d+', '', regex=True)
df['text'] = df['text'].str.lower() # set everything in lowercase

#create a function that splits a string into a list of words,
# then removes the stop words from the list if the words are within NLTKs English stop words list
# finally, concatenates the remaining words back into a string:

def remove_stopwords(text):
  stop = set(stopwords.words('english'))
  text = [word for word in text.split() if word not in stop]
  text = ''.join(x for x in text)
  return text

df['text'] = df['text'].apply(remove_stopwords)

# create stemmer
stemmer = SnowballStemmer("english")

# create a function that splits a string into a list of words
# applies stemmer to each word, 

def string_splitter(text):
  text = [stemmer.stem(word) for word in text.split()]
  text = ''.join(x for x in text)
  return text

df['text'] = df['text'].apply(string_splitter)

df
```
